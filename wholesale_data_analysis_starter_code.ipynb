```python
# Store the client ids of those top 5 clients in a list.
# Source:  https://stackoverflow.com/questions/52441836/how-to-make-a-list-of-values-based-on-value-count-in-pandas
top_five_clients = df['client_id'].value_counts().index[:5].tolist()
print(top_five_clients)

```

```python
# How many total units (the qty column) did the
# client with the most entries order order?
# Part 1 :: Set the client number to a variable
top_client_number = df['client_id'].value_counts().index[:1].tolist()
top_client_number = top_client_number[0]
print(top_client_number)
```

```python
# Part 2 :: Set a new dataframe with topclient only
topclient_df = df.loc[(df["client_id"] == top_client_number)]
topclient_df.head()
```
